```python
import numpy as np
import matplotlib.pyplot as pl
```

```python
# Define parameters:
L=100.
K=0.05  # m/d
```

```python
# Define boundary conditions:
hL=10.
hR=10.
```

```python
# Define driving data:
R=0.001   # m/d
```

```python
# Define x grid:
x=np.linspace(0,L)
```

```python
# Solve:
hs=(hR**2-hL**2)/L*x+R*x/K*(L-x)+hL**2
h=np.sqrt(hs)
```

```python
# Find location of the GW divide:
xGWD=K/2/R*(hR**2-hL**2)/L+L/2
hGWD=(hR**2-hL**2)/L*xGWD+R*xGWD/K*(L-xGWD)+hL**2
hGWD=np.sqrt(hGWD)
```

```python
pl.fill_between([-10,L+10],[-3,-3],0,facecolor='brown')
pl.fill_between([-10,0],[hL,hL],0,facecolor='blue')
pl.fill_between([L,L+10],[hR,hR],0,facecolor='blue')
pl.fill_between([0,L],[15,15],0,facecolor='yellow')
pl.plot(x,h,'-b',linewidth=2)
pl.plot([xGWD,xGWD],[0,hGWD],'--k')
pl.xlabel('Distance, x (m)')
pl.ylabel('Elevation, z (m)')
pl.show()
```
