```python
import numpy as np
import matplotlib.pyplot as pl
```

```python
# Define parameters:
L=100.
b=5.
```

```python
# Define boundary conditions:
hL=10.
hR=11.
```

```python
# Define x grid:
x=np.linspace(0,L)
```

```python
# Solve:
h=(hR-hL)/L*x+hL
```

```python
pl.fill_between([-10,L+10],[-3,-3],0,facecolor='brown')
pl.fill_between([0,L],[12,12],b,facecolor='brown')
pl.fill_between([-10,0],[hL,hL],0,facecolor='blue')
pl.fill_between([L,L+10],[hR,hR],0,facecolor='blue')
pl.fill_between([0,L],[b,b],0,facecolor='yellow')
pl.plot(x,h,'-b',linewidth=2)
pl.xlabel('Distance, x (m)')
pl.ylabel('Elevation, z (m)')
pl.show()
```
